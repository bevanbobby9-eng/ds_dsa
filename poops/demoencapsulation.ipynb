```python
class Calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def sub(self):
        return self.a-self.b
    def mul(self):
        return self.a*self.b
    def div(self):
        return self.a/self.b
        
```

```python
c=Calculator(10,30)
```

```python
c.add()
```

```python
c.sub()
```

```python
c.mul()
```

```python
c.div()
```
